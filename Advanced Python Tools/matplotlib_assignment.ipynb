```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 15, 25, 30, 50])

plt.plot(x,y)
plt.title("simle graph")
plt.xlabel("x-Axis")
plt.ylabel("y-lAxis")
plt.grid(True)
plt.show()
```

```python
students = np.array(['John', 'Jane', 'Alice', 'Bob'])
marks = np.array([75, 85, 60, 90])

plt.bar(students,marks, color = 'orange')
plt.title("Marks of Students")
plt.show()
```

```python
regions = np.array(['North America', 'Europe', 'Asia', 'Others'])
revenue = np.array([45, 25, 20, 10])

indexmax_rev = np.argmax(revenue)
explode = np.zeros(len(revenue))
explode[indexmax_rev] = 0.1

plt.pie(
    revenue,
    labels = regions,
    autopct = '%1.1f%%',
    explode = explode,
    shadow = True,
    startangle= 90,
    )
plt.title("Revenue by Region")
plt.show()
```

```python
data = np.random.randint(1,100, size = 1000)

plt.hist(data)
plt.title("Histogram of Random Data")
plt.xlabel("Value") 
plt.ylabel("Frequency")

plt.show()
```
